```python
df = pd.read_csv("C:\\Users\\saurabh\\Desktop\\Newdat\\attrition.csv")
```

```python
df.head(2) 
```

```python
# df.dtypes
```

```python
df_obj = ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18','OverTime']
```

```python
df= pd.get_dummies(df , columns = df_obj)
```

```python
# df.head(2) 
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
lb = LabelEncoder()
```

```python
df['Attrition'] = lb.fit_transform(df['Attrition'])
```

```python
x = df.drop(columns = ['Attrition'])
y = df['Attrition']
```

```python
# df.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier() 
```

```python
rf.fit(x_train , y_train) 
```

```python
y_pred = rf.predict(x_test) 
```

```python
accuracy_score(y_test , y_pred) 
```
